```python
def funcDraw(x): # This func to make it easy for any plot based on level and subject 
    plt.figure(figsize=(20,10))
    sns.barplot(x='level',y=x,data=df,hue='subject',ci=None)
    plt.title('Plot of {} per level per subject'.format(x))
    plt.show() 
    
    
    plt.figure(figsize=(10,10))
    df.groupby('level')[x].sum().plot(kind='pie')
    plt.title('Plot of {} per level per subject'.format(x))
    plt.show()
```

```python
funcDraw('num_reviews')
```

```python
df['published_date']=df['published_timestamp'].str.split('T').str.get(0) #to seprate date and time and get date only , time is not necessary 
df['published_date']
```

```python
df['published_date']=pd.to_datetime(df['published_date'],format='%Y-%m-%d') #convert time from object to date as data type
df['published_date']
```

```python
df['year']=df['published_date'].dt.year
df['month']=df['published_date'].dt.month
df['day']=df['published_date'].dt.day #seprate year , month and day to 3 columns 
```
